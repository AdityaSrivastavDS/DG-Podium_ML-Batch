```python
#Density Based Spatial Clustering of Applications with Noise (DBSCAN) Algorithm
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Mall_Customers.csv')
X = df.iloc[:, [3, 4]].values
```

```python
X = StandardScaler().fit_transform(X)
```

# DBSCAN Properties:

- eps called as Neighbourhood Radius : How close two points must be to be in the same cluster
- min_samples = Minimum points in cluster : The minimum number of points needed to form a cluster.

```python
dbscan = DBSCAN(eps = 0.5, min_samples = 5)
labels = dbscan.fit_predict(X)
```

```python
#Visualize the cluster
plt.scatter(X[:,0], X[:,1], c = labels, cmap = 'viridis', s = 50)
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
#Finding outliers (Label -1 means outlier)
outliers = X[labels == -1]
print(f"Total Outliers Detected: {len(outliers)}")
```

```python
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis',s=50, label="Clusters")
plt.scatter(outliers[:,0], outliers[:,1], c='red',s=100, label="Outliers", marker='x')
plt.title('DBSCAN Clustering of Customer')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
